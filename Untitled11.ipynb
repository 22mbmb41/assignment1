10.  We may drop the rows with missing values, or we need to input some values into them. We may drop the rows which has missing value/ null value in 'age' column (which has about 20 missing values)

```python
data_sudhir=data_sudhir.dropna(subset=['age'])
```

11.  Let us input values in 'month' column. it is a categorical feature. Hence we can replace the missing values with the MODE value of the column i.e., mostly occurred column. the MODE value you can find out with the function data.month.mode() [in this data is the name of the data frame, month is the name of the column mode() is the function]

```python
data_sudhir['month'].fillna('data_sudhir.month.mode',inplace=True)
```

12.  Display the NULL values again, to confirm that month column does not have any missing values

```python
data_sudhir.isna().sum()
```

13.  'response' column also has missing values.  Since, our target column (DV) is 'response' column, if we impute the values to this column it’ll affect our analysis. So, it is better to drop the missing values from the 'response' column.

```python
data_sudhir=data_sudhir.dropna(subset='response')
```
